```python
import pypsa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#network creation

network = pypsa.Network()

# snapshots (8760 hourly steps)
#network.set_snapshots(pd.date_range("2023-01-01", periods=8760, freq="h"))
network.set_snapshots(range(8760))
#subsystem 

network.add("Bus", name= "e_bus") 
network.add("Bus", name= "h_bus")

# generator with operating cost
network.add("Generator", name= "PV", bus= "e_bus", p_nom= 50, marginal_cost=0.01)

# heat pump link with investment cost (extendable)
network.add("Link", name= "hetapump", bus0= "e_bus", bus1= "h_bus", efficiency= 1.00, p_nom_extendable=True)

# heat load as time series
network.add("Load", name= "h_load", bus= "h_bus", p_set= pd.Series(10, index=network.snapshots))




```

```python
network.optimize(solver_name='highs')
```
